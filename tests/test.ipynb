```python
import pandas
```

```python
test_1 = {"true_negative": {
    "validation__detection": 118,
    "validation__liveness": 110,
    "validation__suitability": 75,
    "authentication": 193,
}}
test_2 = {"false_positive": 4}
test_3 = {"false_negative": {
    "validation__detection": 63,
    "validation__liveness": 37,
    "validation__suitability": 22,
    "authentication": 69,
}}
test_4 = {"true_positive": 309}
```

```python
df1 = pandas.DataFrame.from_dict(test_1)
df2 = pandas.DataFrame(test_2, index=[0])
df3 = pandas.DataFrame.from_dict(test_3)
df4 = pandas.DataFrame(test_4, index=[0])
```

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [11, 5]
_, axes = plt.subplots(nrows=1, ncols=2)

ax = dff = pandas.concat([df1, df2]).plot.bar(ax=axes[0])
dft = pandas.concat([df4, df3]).plot.bar(ax=axes[1])
```
